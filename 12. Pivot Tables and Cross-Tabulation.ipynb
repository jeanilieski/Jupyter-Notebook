```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

# very good site: http://pbpython.com/pandas-pivot-table-explained.html
```

```python
tips=pd.read_csv('tips.csv');

# add col 'tip percent of total bill'
tips['tip_pct']=tips['tip']/tips['total_bill'];

tips[:6]
```

```python
# to compute a table of group means(the default pivot_table 
# aggregation type) arranged by sex and smoker:

tips.pivot_table(index=['sex','smoker'])
```

```python
# if interested only in some cols you may specif them in 'values':

tips.pivot_table(index=['sex','smoker'], values=['tip_pct', 'size'])
```

```python
# aggr only 'size' and 'tip_pct',  additionally group by 'day', 
# put 'smoker' into columns

tips.pivot_table(['tip_pct', 'size'], index=['sex', 'day'], 
                columns='smoker')
```

```python
#to see the absolute numbers behind the means:

tips.pivot_table(['tip_pct', 'size'], index=['sex', 'day'], 
                columns='smoker', aggfunc='count', margins=True)

# you can also use aggfunc=len insteda of aggfunc='count'
# margins=True yields the 'All' cols and rows
```

```python
#if you need sum and the 'count' insted of the default operation 'mean':

tips.pivot_table(['tip'], index=['sex', 'smoker'], 
                aggfunc=[np.sum,len], margins=True)
```

```python
#you can create a fixed table and then filter it:

table=tips.pivot_table(['tip_pct', 'size'], index=['sex', 'day'], 
                columns='smoker', aggfunc='count', margins=True);

table
```

```python
table.query('sex == ["Female"]') 

# must use "" (not '')in this case for Female
```

```python
table.query('day==["Fri", "Sat"]')

# you can filter only by rows, so: table.query('smoker==["No"]')
# would not function, even if the 'All' col does not exist
```

```python
# Cross-Tabulations: CROSSTAB

#mostly used for crossing of two categorical variables 
```

```python
df = DataFrame({
'key1' : ['a', 'a', 'b', 'b', 'a'],
'key2' : ['one', 'two', 'one', 'two', 'one'],
'data1' : [2,4,6,3,5],
'data2' : [8,5,9,1,4]});

df
```

```python
pd.crosstab(df.key1, df.key2, margins=True)
```

```python
pd.crosstab(df.key1, df.key2).apply(lambda r:r/r.sum(), axis=1) #percentage crosstab is also possible
```

```python
pd.crosstab([tips.time, tips.day], tips.smoker, margins=True)
```

```python
# another example with percentage:

df = pd.DataFrame({'cow' : ['one', 'one', 'two', 'three'] * 6,
                   'bull' : ['A', 'B', 'C', 'D', 'C', 'D'] * 4,
                   'C' : ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 4,
                   'D' : np.random.randn(24),
                   'E' : np.random.randn(24)});
df
```

```python
pd.crosstab(df.cow,df.bull, margins=True)
```

```python
pd.crosstab(df.cow, df.bull).apply(lambda r: r/r.sum(), axis=1)
```
