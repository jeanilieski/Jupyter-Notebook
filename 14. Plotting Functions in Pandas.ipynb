```python
# from docs, e.g, CSV:

tips=pd.read_csv('tips.csv')
```

```python
gen_counts=pd.crosstab(tips.sex, tips.day);
gen_counts
```

```python
# Normalize to sum to 1
gen_pcts = gen_counts.div(gen_counts.sum(1).astype(float), axis=0)
```

```python
gen_pcts
```

```python
gen_pcts.plot(kind='bar', stacked=True)
# Male come more on Sunday and Female on Thursday
```

```python
# HISTOGRAM AND DENSITY PLOTS:

tips['tip_pct']=tips['tip']/tips['total_bill'];
tips['tip_pct'].mean()
```

```python
tips['tip_pct'].hist(bins=50)
```
