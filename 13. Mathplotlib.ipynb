```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)

# sharex - all the subplots will have the same X-axis: this 
# makes easier the comparison of data on same scales
# alpha=0.5 is the brightness of color

#plt.subplots_adjust(left=None, bottom=None, right=None, top=None,
               #wspace=None, hspace=None)
```

```python
# costumising axis:

fig3=plt.figure(); 
ax=fig3.add_subplot(1,1,1); 
ax.plot(randn(1000).cumsum())
```

```python
ticks=ax.set_xticks([0,250,500,750,1000]); # change ticks 
fig3
```

```python
labels=ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], 
                          fontsize='small'); # set labels to ticks
ax.set_title('Cumulative Random Walk'); # set title to chart
ax.set_xlabel('Stages'); # set label to axis

fig3
```

```python
fig4=plt.figure(); #adding legends
ax=fig4.add_subplot(1,1,1); 
ax.plot(randn(100).cumsum(), 'k', label='one');
ax.plot(randn(100).cumsum(), 'k--', label='two');
ax.plot(randn(100).cumsum(), 'k.', label='three')
```

```python
ax.legend(loc='best'); # location=best
fig4
```
